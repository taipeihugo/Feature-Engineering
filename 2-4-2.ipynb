```python
import pandas as pd
# 參考資料: https://pandas.pydata.org/docs/user_guide/merging.html
# Create two sample DataFrames
df1 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30]})
df2 = pd.DataFrame({'Name': ['Charlie', 'David'], 'Age': [35, 40]})

# Combine using concat()
combined_df = pd.concat([df1, df2])
display(combined_df)
```

```python
combined_df = pd.concat([df1, df2], ignore_index=True)
display(combined_df)
```

```python
df1 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30]})
df2 = pd.DataFrame({'City': ['New York', 'Los Angeles'], 'Salary': [70000, 80000]})

# Combine using concat() along columns
combined_df = pd.concat([df1, df2], axis=1, ignore_index=False)
display(combined_df)
```

```python
import pandas as pd
# Merging on a common column
df1 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30]})
df2 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Salary': [50000, 60000]})

merged_df = pd.merge(df1, df2, on='Name')
display(merged_df)
```

```python
left = pd.DataFrame(
    {"A": ["A0", "A1", "A2"], "B": ["B0", "B1", "B2"]}, index=["K0", "K1", "K2"]
)
left
```

```python
right = pd.DataFrame(
    {"C": ["C0", "C2", "C3"], "D": ["D0", "D2", "D3"]}, index=["K0", "K2", "K3"]
)
right
```

```python
result = left.join(right)
result
```

```python
result = left.join(right, how="outer")
result
```

```python
result = left.join(right, how="inner")
result
```

```python
left = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "key": ["K0", "K1", "K0", "K1"],
    }
)
left
```

```python
right = pd.DataFrame({"C": ["C0", "C1"], "D": ["D0", "D1"]}, index=["K0", "K1"])
right
```

```python
result = left.join(right, on="key")
result
```
